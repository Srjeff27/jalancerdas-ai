# 🏗️ Build APK - JalanCerdas AI

Notebook ini membangun APK release dari project Flutter menggunakan Colab.

**Langkah:**
1. Clone project dari GitHub
2. Install Flutter SDK + Android SDK
3. Build APK release
4. Download APK hasil build

## 📋 Cell 1: Install Java 17 & Flutter SDK

```python
# Install Java 17 (required by Gradle)
!sudo apt-get update -qq
!sudo apt-get install -y -qq openjdk-17-jdk > /dev/null 2>&1
print('✅ Java 17 installed')

# Set JAVA_HOME
import os
os.environ['JAVA_HOME'] = '/usr/lib/jvm/java-17-openjdk-amd64'
print(f'JAVA_HOME: {os.environ["JAVA_HOME"]}')

# Verify Java version
!java -version
```

## 📋 Cell 2: Clone Project & Install Flutter

```python
# Clone project
!git clone https://github.com/Srjeff27/jalancerdas-ai.git
%cd jalancerdas-ai/mobile

# Install Flutter SDK
!git clone https://github.com/flutter/flutter.git -b stable --depth 1 > /dev/null 2>&1

# Set PATH
import os
os.environ['PATH'] = f"/content/jalancerdas-ai/mobile/flutter/bin:{os.environ['PATH']}"
os.environ['FLUTTER_ROOT'] = '/content/jalancerdas-ai/mobile/flutter'

# Flutter doctor
!flutter doctor -v 2>&1 | head -30
```

## 📋 Cell 3: Install Android SDK & Accept Licenses

```python
# Setup Android SDK
import os

ANDROID_HOME = '/opt/android-sdk'
os.environ['ANDROID_HOME'] = ANDROID_HOME
os.environ['ANDROID_SDK_ROOT'] = ANDROID_HOME

!mkdir -p $ANDROID_HOME/cmdline-tools

# Download Android command-line tools
!wget -q https://dl.google.com/android/repository/commandlinetools-linux-11076708_latest.zip -O /tmp/cmdline-tools.zip
!unzip -q /tmp/cmdline-tools.zip -d /tmp/cmdline-tools-tmp
!mv /tmp/cmdline-tools-tmp/cmdline-tools $ANDROID_HOME/cmdline-tools/latest

# Add to PATH
os.environ['PATH'] = f"{ANDROID_HOME}/cmdline-tools/latest/bin:{ANDROID_HOME}/platform-tools:{os.environ['PATH']}"

# Accept all licenses
!yes | sdkmanager --licenses > /dev/null 2>&1

# Install required SDK components
!sdkmanager --install "platforms;android-35" "build-tools;35.0.0" "ndk;28.2.13676358" "platform-tools" 2>&1 | tail -5

print('✅ Android SDK installed')
```

## 📋 Cell 4: Configure local.properties

```python
# Create local.properties
with open('android/local.properties', 'w') as f:
    f.write(f"""sdk.dir={ANDROID_HOME}
flutter.sdk=/content/jalancerdas-ai/mobile/flutter
flutter.buildMode=release
flutter.versionName=1.0.0
flutter.versionCode=1
""")

print('✅ local.properties created')
!cat android/local.properties
```

## 📋 Cell 5: Build APK Release 🚀

```python
# Build APK
!flutter pub get
!flutter build apk --release

# Check output
!ls -lh build/app/outputs/flutter-apk/
```

## 📋 Cell 6: Download APK 📥

```python
from google.colab import files

apk_path = 'build/app/outputs/flutter-apk/app-release.apk'

import os
if os.path.exists(apk_path):
    print(f'✅ APK ready! Size: {os.path.getsize(apk_path) / 1024 / 1024:.1f} MB')
    files.download(apk_path)
else:
    print('❌ APK not found. Check build output above.')
```
